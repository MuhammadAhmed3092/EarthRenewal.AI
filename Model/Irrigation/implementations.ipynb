```python
import pandas as pd
file_path = '/content/LIRF Maize 2011 r1_0_.xlsx'
df = pd.read_excel(file_path, sheet_name='Tmnt4')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Precip\nmm'].fillna(df['Precip\nmm'].mean(), inplace=True)
df['Irrig\nmm'].fillna(df['Irrig\nmm'].mean(), inplace=True)
df['0 - 15\n%'].fillna(df['0 - 15\n%'].median(), inplace=True)
df['15 - 45\n%'].fillna(df['15 - 45\n%'].median(), inplace=True)
df['45 - 75\n%'].fillna(df['45 - 75\n%'].median(), inplace=True)
df['75 - 105\n%'].fillna(df['75 - 105\n%'].median(), inplace=True)
df['105 - 135\n%'].fillna(df['105 - 135\n%'].median(), inplace=True)
df['135 - 165\n%'].fillna(df['135 - 165\n%'].median(), inplace=True)
df['165 - 200\n%'].fillna(df['165 - 200\n%'].median(), inplace=True)
df['Growth\nStage'].fillna(df['Growth\nStage'].mode()[0], inplace=True)
df['Root\nDepth'].fillna(df['Root\nDepth'].mean(), inplace=True)
df['Canopy\nCover'].fillna(df['Canopy\nCover'].mean(), inplace=True)
df['Height\ncm'].fillna(df['Height\ncm'].mean(), inplace=True)
df['N Applied\nkg/ha'].fillna(df['N Applied\nkg/ha'].mean(), inplace=True)
df['ETr\nmm'].fillna(df['ETr\nmm'].mean(), inplace=True)
df['Kcb'].fillna(df['Kcb'].mean(), inplace=True)
df['Ks'].fillna(df['Ks'].mean(), inplace=True)
df['ETcb\nmm'].fillna(df['ETcb\nmm'].mean(), inplace=True)
df['Evap\nmm'].fillna(df['Evap\nmm'].mean(), inplace=True)
df['Deep Perc\nmm'].fillna(df['Deep Perc\nmm'].mean(), inplace=True)
df['ETc\nmm'].fillna(df['ETc\nmm'].mean(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.head()
```

```python
df_train = df.drop(['DOY', 'Active Root Zone\nPredicted'], axis = 1)
```

```python
df_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df_train['Growth\nStage'] = encoder.fit_transform(df['Growth\nStage'])
```

```python
df_train.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for i in df_train.columns:
    plt.title("Distribution of {i}")
    sns.histplot(x = i, data = df_train, kde = True , bins = 5)
    plt.show()
```

```python
df['Active Root Zone\nPredicted'].plot()
```

```python
df.loc[0, 'Active Root Zone\nPredicted'] = df['Active Root Zone\nPredicted'].mean()

```

```python
df['Active Root Zone\nPredicted'].isna().sum()
```

```python
x = df_train
y = df['Active Root Zone\nPredicted']
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2, random_state = 3)
```

```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(xtrain, ytrain)
print('training result : ', linear.score(xtrain, ytrain))
print("testing result : ", linear.score(xtest, ytest))
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

y_pred = linear.predict(xtest)

# Mean Absolute Error
mae = mean_absolute_error(ytest, y_pred)
print(f'Mean Absolute Error: {mae}')

# R-squared
r2 = r2_score(ytest, y_pred)
print(f'R-squared: {r2}')
```

# applying cnn


```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv1D, Flatten, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
x_train_cnn = np.expand_dims(xtrain, axis=-1)
x_test_cnn = np.expand_dims(xtest, axis=-1)
```

```python
cnn_model = Sequential()
cnn_model.add(Conv1D(filters=64, kernel_size=2, activation='relu', input_shape=(x_train_cnn.shape[1], x_train_cnn.shape[2])))
cnn_model.add(Dropout(0.5))
cnn_model.add(Conv1D(filters=64, kernel_size=2, activation='relu'))
cnn_model.add(Flatten())
cnn_model.add(Dense(50, activation='relu'))
cnn_model.add(Dense(1))
```

```python
cnn_model.compile(optimizer=Adam(learning_rate=0.01), loss='mean_squared_error')
```

```python
cnn_history = cnn_model.fit(x_train_cnn, ytrain, epochs=100, batch_size=32, validation_data=(x_test_cnn, ytest))
```

```python
# Evaluate CNN
cnn_eval = cnn_model.evaluate(x_test_cnn, ytest, verbose=0)
print(f'CNN Mean Squared Error: {cnn_eval}')

```

```python
from sklearn.metrics import mean_absolute_error, r2_score

y_pred = cnn_model.predict(x_test_cnn)

# Mean Absolute Error
mae = mean_absolute_error(ytest, y_pred)
print(f'Mean Absolute Error: {mae}')

# R-squared
r2 = r2_score(ytest, y_pred)
print(f'R-squared: {r2}')
```

```python
from tensorflow.keras.layers import BatchNormalization

# Define an updated CNN model
cnn_model = Sequential()
cnn_model.add(Conv1D(filters=128, kernel_size=3, activation='relu', input_shape=(x_train_cnn.shape[1], x_train_cnn.shape[2])))
cnn_model.add(BatchNormalization())
cnn_model.add(Dropout(0.5))
cnn_model.add(Conv1D(filters=64, kernel_size=3, activation='relu'))
cnn_model.add(Flatten())
cnn_model.add(Dense(100, activation='relu'))
cnn_model.add(Dense(1))

# Compile the model
cnn_model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')

# Train the model
cnn_history = cnn_model.fit(x_train_cnn, ytrain, epochs=150, batch_size=32, validation_data=(x_test_cnn, ytest))

# Evaluate the CNN model
cnn_eval = cnn_model.evaluate(x_test_cnn, ytest, verbose=0)
print(f'Updated CNN Mean Squared Error: {cnn_eval}')

# Predict on the test set
y_pred = cnn_model.predict(x_test_cnn)

# Mean Absolute Error
mae = mean_absolute_error(ytest, y_pred)
print(f'Updated Mean Absolute Error: {mae}')

# R-squared
r2 = r2_score(ytest, y_pred)
print(f'Updated R-squared: {r2}')
```

```python
import numpy as np
import torch
from sklearn.metrics import mean_absolute_error, r2_score
from pytorch_tabnet.tab_model import TabNetRegressor

xtrain_np = xtrain.values.astype(np.float32)
ytrain_np = ytrain.values.astype(np.float32).reshape(-1, 1)
xtest_np = xtest.values.astype(np.float32)
ytest_np = ytest.values.astype(np.float32).reshape(-1, 1)

# Initialize TabNetRegressor
model = TabNetRegressor()

# Train the model
model.fit(
    X_train=xtrain_np,
    y_train=ytrain_np,
    max_epochs=100,
    patience=10,
    batch_size=32,
    virtual_batch_size=16,
    num_workers=0,
    drop_last=False
)

# Predict on the test set
preds = model.predict(xtest_np)

# Calculate additional metrics
mae = mean_absolute_error(ytest_np, preds)
r2 = r2_score(ytest_np, preds)

print(f'Mean Absolute Error: {mae}')
print(f'R-squared: {r2}')
```
